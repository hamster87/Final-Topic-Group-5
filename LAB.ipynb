```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the images
s0 = cv2.imread("s0.jpg")
s4 = cv2.imread("view_1.jpeg")

# Convert images to RGB
s_0 = cv2.cvtColor(s0, cv2.COLOR_BGR2RGB)
s_4 = cv2.cvtColor(s4, cv2.COLOR_BGR2RGB)

# Convert images to Lab color space and extract the L channel
s_0_lab = cv2.cvtColor(s_0, cv2.COLOR_RGB2Lab)
s_4_lab = cv2.cvtColor(s_4, cv2.COLOR_RGB2Lab)

L_0 = s_0_lab[:, :, 0]
L_4 = s_4_lab[:, :, 0]

# Display the L channels
query_images = [L_0, L_4]
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(query_images[i], cmap='gray')
    plt.xticks([]), plt.yticks([])

plt.show()

# Compute histograms for the L channels
hist_0, bins_0 = np.histogram(L_0.flatten(), bins=256, range=[0, 256])
hist_4, bins_4 = np.histogram(L_4.flatten(), bins=256, range=[0, 256])

# Normalize the histograms
hist_0 = hist_0 / np.sum(hist_0)
hist_4 = hist_4 / np.sum(hist_4)

# Plot the histograms
plt.figure()
plt.plot(hist_0, color='r', label='s0 - L channel')
plt.plot(hist_4, color='b', label='view_1 - L channel')
plt.xlabel('Pixel value')
plt.ylabel('Normalized frequency')
plt.legend()
plt.title('L channel histograms')
plt.show()
```
